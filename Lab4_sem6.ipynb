```python
import datetime
import matplotlib.pyplot as plt

# Создаем список сотрудников
employees = [
    {
        "№": 1,
        "Таб. номер": "001",
        "Фамилия": "Иванов",
        "Дата поступления": datetime.date(2015, 5, 12),
        "Дата рождения": datetime.date(1985, 8, 20),
        "Оклад": 50000
    },
    {
        "№": 2,
        "Таб. номер": "002",
        "Фамилия": "Петрова",
        "Дата поступления": datetime.date(2018, 3, 15),
        "Дата рождения": datetime.date(1990, 11, 5),
        "Оклад": 45000
    },
    {
        "№": 3,
        "Таб. номер": "003",
        "Фамилия": "Сидоров",
        "Дата поступления": datetime.date(2020, 7, 1),
        "Дата рождения": datetime.date(1995, 4, 30),
        "Оклад": 40000
    },
    {
        "№": 4,
        "Таб. номер": "004",
        "Фамилия": "Кузнецова",
        "Дата поступления": datetime.date(2010, 9, 22),
        "Дата рождения": datetime.date(1980, 12, 10),
        "Оклад": 60000
    },
    {
        "№": 5,
        "Таб. номер": "005",
        "Фамилия": "Смирнов",
        "Дата поступления": datetime.date(2019, 2, 14),
        "Дата рождения": datetime.date(1992, 7, 25),
        "Оклад": 48000
    },
    {
        "№": 6,
        "Таб. номер": "006",
        "Фамилия": "Васильева",
        "Дата поступления": datetime.date(2017, 6, 30),
        "Дата рождения": datetime.date(1988, 3, 8),
        "Оклад": 52000
    },
    {
        "№": 7,
        "Таб. номер": "007",
        "Фамилия": "Попов",
        "Дата поступления": datetime.date(2016, 11, 9),
        "Дата рождения": datetime.date(1983, 9, 17),
        "Оклад": 55000
    },
    {
        "№": 8,
        "Таб. номер": "008",
        "Фамилия": "Федорова",
        "Дата поступления": datetime.date(2021, 4, 5),
        "Дата рождения": datetime.date(1999, 1, 12),
        "Оклад": 42000
    }
]

# Функция для вычисления возраста
def calculate_age(birth_date):
    today = datetime.date.today()
    return today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))

# Функция для вычисления стажа
def calculate_experience(start_date):
    today = datetime.date.today()
    return today.year - start_date.year - ((today.month, today.day) < (start_date.month, start_date.day))

# Добавляем вычисляемые поля
for emp in employees:
    emp["Возраст"] = calculate_age(emp["Дата рождения"])
    emp["Стаж"] = calculate_experience(emp["Дата поступления"])

# Вывод данных по табельному номеру
tab_number = input("Введите табельный номер сотрудника (например, 001): ")
found = False
for emp in employees:
    if emp["Таб. номер"] == tab_number:
        print("\nДанные сотрудника:")
        print(f"Табельный номер: {emp['Таб. номер']}")
        print(f"Фамилия: {emp['Фамилия']}")
        print(f"Дата поступления: {emp['Дата поступления']}")
        print(f"Дата рождения: {emp['Дата рождения']}")
        print(f"Оклад: {emp['Оклад']} руб.")
        print(f"Возраст: {emp['Возраст']} лет")
        print(f"Стаж: {emp['Стаж']} лет")
        found = True
        break
if not found:
    print("Сотрудник с таким табельным номером не найден.")

# Построение диаграммы распределения по возрасту
ages = [emp["Возраст"] for emp in employees]
plt.hist(ages, bins=5, edgecolor='black')
plt.title('Распределение сотрудников по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Количество сотрудников')
plt.grid(True)
plt.show()
```

### Описание работы кода:
1. **Создание списка сотрудников**: В списке `employees` хранятся словари с данными 8 сотрудников.
2. **Функции для вычисления возраста и стажа**: 
   - `calculate_age` вычисляет возраст на основе даты рождения.
   - `calculate_experience` вычисляет стаж работы на основе даты поступления.
3. **Добавление вычисляемых полей**: Для каждого сотрудника добавляются поля "Возраст" и "Стаж".
4. **Поиск сотрудника по табельному номеру**: Пользователь вводит табельный номер, программа выводит данные сотрудника или сообщение об ошибке.
5. **Построение диаграммы**: Гистограмма показывает распределение сотрудников по возрастным группам.
